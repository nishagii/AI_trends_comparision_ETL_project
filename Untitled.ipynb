```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("ai_trends.csv")

# Convert the date column to datetime
data["date"]=pd.to_datetime(data["date"])

data.fillna(0,inplace=True)

# Add new columns for Year and Month
data['Year'] = data['date'].dt.year
data['Month'] = data['date'].dt.month
```

```python
# View the entire DataFrame
print(data)
```

```python
df_ai_trends=pd.DataFrame(data)
print(df_ai_trends)
```

```python
#monthly average popularity

monthly_data=data.groupby(['Year','Month']).mean().reset_index()
```

```python
print(monthly_data.columns)
```

```python
plt.plot(monthly_data['Year'].astype(str) + '-' + monthly_data['Month'].astype(str), monthly_data['chatGPT'])
plt.xticks(ticks=range(0, len(monthly_data), 6), rotation=45)  # Show every 6th month
plt.title('ChatGPT Popularity Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Popularity')
plt.tight_layout()
plt.show()
```

```python
plt.plot(monthly_data['Year'].astype(str) + '-' + monthly_data['Month'].astype(str), monthly_data['Generative AI'])
plt.xticks(ticks=range(0, len(monthly_data), 6), rotation=45)  # Show every 6th month
plt.title('Generative AI Popularity Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Popularity')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Prepare the data
prophet_data = merged_data[['date', 'chatGPT']].rename(columns={'date': 'ds', 'chatGPT': 'y'})

# Ensure the 'ds' column is in datetime format
prophet_data['ds'] = pd.to_datetime(prophet_data['ds'])

# Set the 'ds' column as the index
prophet_data.set_index('ds', inplace=True)

# Plot ACF and PACF to determine p and q
plot_acf(prophet_data['y'])
plot_pacf(prophet_data['y'])
plt.show()

# Fit ARIMA model (using p, d, q values determined from ACF/PACF)
# Example: ARIMA(5,1,0)
model = ARIMA(prophet_data['y'], order=(5, 1, 0))
model_fit = model.fit()

# Make predictions (forecast the next 365 days)
forecast = model_fit.forecast(steps=365)

# Plot forecasted values
plt.figure(figsize=(10,6))
plt.plot(prophet_data.index, prophet_data['y'], label='Historical Data')
plt.plot(pd.date_range(prophet_data.index[-1], periods=366, freq='D')[1:], forecast, label='Forecast', color='red')
plt.legend()
plt.show()

# Forecast values
print(forecast)
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit the model
model = ExponentialSmoothing(prophet_data['y'], trend='add', seasonal='add', seasonal_periods=12)
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=365)

# Plot forecasted values
plt.plot(prophet_data.index, prophet_data['y'], label='Historical Data')
plt.plot(pd.date_range(prophet_data.index[-1], periods=366, freq='D')[1:], forecast, label='Forecast', color='red')
plt.legend()
plt.show()
```
